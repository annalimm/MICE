```python
def metr_func(X_miss, imp_name, df_tr, metrica, col2, df_fin, scenar):
    meth_imp = Imputers.impute(X_miss, imp_name)  
    res_X, res_col2 = Metrics.metr(df_tr, X_miss, imp_name, meth_imp, metrica, col2, scenar)


    df_fin = pd.concat([df_fin, 
                      pd.DataFrame.from_records([{'estimator':imp_name, 'col':col2, metrica: res_col2}])])
    
    return df_fin
```

```python
def imp(metrica, varss, col2): 
    
    df_fin = pd.DataFrame(columns = ['estimator', 'col', metrica])

    for n_c in tqdm(np.arange(30)):

        
        #df = gener_dt.gener(X_distr = dist, scenar = scenar)
        #X_miss = HG_tools.amp(df)
        #X_miss = HG_tools.amp_MNAR(df)
        
        scenar = 'lr_quadr' #пусть так
        
        
        ma = MultivariateAmputation( #, 'weights': weis
            patterns = [ {'incomplete_vars': varss, 'mechanism': 'MAR', 'score_to_probability_func': 'sigmoid-right'} ],
            prop = 0.45
            )
        X_miss = ma.fit_transform(df_dropna)
        
        
        print('mean')
        df_fin = metr_func(X_miss, 'mean', df_dropna, metrica, col2, df_fin, scenar)
        
        print('mf')
        df_fin = metr_func(X_miss, 'mf', df_dropna, metrica, col2, df_fin, scenar)
        
        #miceRF
        #---------------------------
        print('miceRF')
        df_fin = metr_func(X_miss, 'mice_rf', df_dropna, metrica, col2, df_fin, scenar)

    
        
        #miceBR
        #---------------------------
        #print('mice_br')
        #df_fin = metr_func(X_miss, 'mice_br', df_dropna, metrica, col2, df_fin, scenar)

        #knn
        #---------------------------
        print('knn')
        df_fin = metr_func(X_miss, 'knn', df_dropna, metrica, col2, df_fin, scenar)



    return df_fin
```

```python
def plott(res, metrica, col, xlim_from, xlim_to):
    plt.figure(figsize =  (4, 3.5))

    sns.boxplot(x = metrica, y = 'estimator', 
                data = res[res['col'] == col], 
                palette = "colorblind",
                hue = 'estimator')

    plt.grid(alpha = 0.5, linestyle = '--', linewidth = 0.6)
    # plt.title("Ivan." + str(metrica) + " by " + str(col) )
    #plt.title("MAR, 0.45. "+ str(metrica) + " by " + str(col) )
    #plt.title(str(metrica))
    plt.tight_layout()
    plt.xlim(xlim_from, xlim_to)
    #plt.show()
    
    plt.savefig('/Users/ash/Desktop/MICE/PY/Iv_nrmse.png', format = 'png',transparent=True, dpi=300)
```
