```python
distrs = ['norm', 'unif', 'lognorm1', 'lognorm2', 'gamma1', 'gamma2', 'norm_mix1', 'norm_mix2']
```

```python
def df_mean_col(imp_mice, col):
    m_means = []
    
    for imp in imp_mice:
        
        imp_mean = pd.DataFrame(imp).loc[:, col].mean()
        m_means.append(imp_mean)
        

    fin_min = np.array(m_means).mean()

    return fin_min


def amp(df):
    if len(df.columns) == 3:
        wei = [0, 1, 0]
    else:
        wei = [0, 1, 0, 0]
    
    ma = MultivariateAmputation(
                patterns = [ {'incomplete_vars': [0, 2], 'weights': wei, 'mechanism': 'MAR', 'score_to_probability_func': 'sigmoid-right'} ],
                prop = 0.25
                )

    df_miss_MAR = ma.fit_transform(df)
    return df_miss_MAR
```

```python
def imp_bias(n_cycle, scenar, col2): 
    
    df_bias_imp = pd.DataFrame(columns = ['dist', 'estimator', 'col', 'rel_bias'])

    for dist in tqdm(distrs):
        print(dist)
        for n_c in tqdm(np.arange(n_cycle)):

            df = gener_dt.gener(X_distr = dist, scenar = scenar)
            df_X_mean = df.loc[:, 'X'].mean()
            df_col2_mean = df.loc[:, col2].mean()
            X_miss = amp(df)
            
            #iceRF
            ICE_RF_imp = Imputers.impute(X_miss, imputer_name = 'ice_rf')
            ICE_RF_X_mean = ICE_RF_imp.loc[:, 'X'].mean()
            ICE_RF_X_rel_bias = (ICE_RF_X_mean / df_X_mean) - 1
            
            ICE_RF_col2_mean = ICE_RF_imp.loc[:, col2].mean()
            ICE_RF_col2_rel_bias = (ICE_RF_col2_mean / df_col2_mean) - 1
            
            
            #miceRF
            MICE_RF_imp = Imputers.impute(X_miss, imputer_name = 'mice_rf')
            MICE_RF_X_mean = df_mean_col( MICE_RF_imp, 'X')
            MICE_RF_X_rel_bias = (MICE_RF_X_mean / df_X_mean) - 1
            
            MICE_RF_col2_mean = df_mean_col( MICE_RF_imp, col2 )
            MICE_RF_col2_rel_bias = (MICE_RF_col2_mean / df_col2_mean) - 1
            
            
            #miceBR
            MICE_BR_imp = Imputers.impute(X_miss, imputer_name = 'mice_br')
            MICE_BR_X_mean = df_mean_col( MICE_BR_imp, 'X')
            MICE_BR_X_rel_bias = (MICE_BR_X_mean / df_X_mean) - 1
            
            MICE_BR_col2_mean = df_mean_col( MICE_BR_imp, col2)
            MICE_BR_col2_rel_bias = (MICE_BR_col2_mean / df_col2_mean) - 1
            
            #knn
            knn_imp = Imputers.impute(X_miss, imputer_name = 'knn')
            knn_X_mean = knn_imp.loc[:, 'X'].mean()
            knn_X_rel_bias = (knn_X_mean / df_X_mean) - 1
            
            knn_col2_mean = knn_imp.loc[:, col2].mean()
            knn_col2_rel_bias = (knn_col2_mean / df_col2_mean) - 1

            
            df_bias_imp = pd.concat([df_bias_imp, 
                                     pd.DataFrame.from_records([{'dist':dist, 'estimator':'ice_rf', 'col':'X', 'rel_bias':ICE_RF_X_rel_bias}]),
                                     pd.DataFrame.from_records([{'dist':dist, 'estimator':'ice_rf', 'col':col2, 'rel_bias':ICE_RF_col2_rel_bias}]),
                                     pd.DataFrame.from_records([{'dist':dist, 'estimator':'mice_rf', 'col':'X', 'rel_bias':MICE_RF_X_rel_bias}]),
                                     pd.DataFrame.from_records([{'dist':dist, 'estimator':'mice_rf', 'col':col2, 'rel_bias':MICE_RF_col2_rel_bias}]),
                                     pd.DataFrame.from_records([{'dist':dist, 'estimator':'knn', 'col':'X', 'rel_bias':knn_X_rel_bias}]),
                                     pd.DataFrame.from_records([{'dist':dist, 'estimator':'knn', 'col':col2, 'rel_bias':knn_col2_rel_bias}]),
                                     pd.DataFrame.from_records([{'dist':dist, 'estimator':'mice_br', 'col':'X', 'rel_bias':MICE_BR_X_rel_bias}]),
                                     pd.DataFrame.from_records([{'dist':dist, 'estimator':'mice_br', 'col':col2, 'rel_bias':MICE_BR_col2_rel_bias}])
                                    ])



    return df_bias_imp
```
